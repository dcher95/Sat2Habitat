```python
import pickle
import pandas as pd
```

```python

# Define experiments
experiments = [
    # 'full-data-clip', 
    "full-data-crisp250-nowiki",
    "full-data-crisp0.5",
    "full-data-crisp0.8",
    "full-data-crisp250-0.9-exp",
    "full-data-crisp+0.9+exp",
    # "full-data-crisp250-0.9-exp",
]


# "sat2hab-crisp"
```

```python

for experiment in experiments:
    metrics_path = f"/data/cher/Sat2Habitat/metrics/metrics_{experiment}.pkl"
    metrics_county_path = f"/data/cher/Sat2Habitat/metrics/metrics_{experiment}_county.pkl"
    metrics_eco_path = f"/data/cher/Sat2Habitat/metrics/metrics_{experiment}_eco.pkl"

    with open(metrics_path, 'rb') as f:
        metrics = pickle.load(f)
    metrics['accuracy']

    with open(metrics_county_path, 'rb') as f:
        metrics_county = pickle.load(f)
    metrics_county['accuracy']

    with open(metrics_eco_path, 'rb') as f:
        metrics_eco = pickle.load(f)
    metrics_eco['accuracy']
```

```python
# Initialize lists to store results
experiment_names = []
recalls = []

# Iterate through each experiment
for experiment in experiments:
    metrics_path = f"/data/cher/Sat2Habitat/metrics/metrics_{experiment}.pkl"
    with open(metrics_path, 'rb') as f:
        metrics = pickle.load(f)
    
    # Collect recall values for this experiment
    recalls.append(metrics['accuracy'])
    experiment_names.append(experiment)

# Create a DataFrame
df = pd.DataFrame(recalls, index=experiment_names)
df.index.name = "Experiment"
df.columns = ["Recall@1", "Recall@5", "Recall@10", "Recall@25"]

# Display the DataFrame
display(df)
```
