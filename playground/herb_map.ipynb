```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
```

```python
# location information & occurrence info
occ = pd.read_csv("/data/cher/Sat2Habitat/data/occurrence.txt", sep="\t", on_bad_lines='skip')
occ = occ[[
    'gbifID', 'occurrenceID', # Joining Keys
    'speciesKey', 'species', # species
    'year', 'month', 'day', # date
    'habitat', 'locality', # descriptive text. I think this is most of the route / habitat info we would want. Looks like already OCR processed?
    'countryCode', 'stateProvince', 'county', 'municipality', # administrative
    'georeferenceSources','hasCoordinate', 'hasGeospatialIssues' ,'decimalLatitude', 'decimalLongitude']] # geospatial
    # 'level0Gid', 'level0Name', 'level1Gid', 'level1Name', 'level2Gid', 'level2Name', 'level3Gid', 'level3Name',]] # Has NAs. Above is more reliable.
```

```python
occ['decimalLatitude'] = pd.to_numeric(occ['decimalLatitude'], errors='coerce')
occ['decimalLongitude'] = pd.to_numeric(occ['decimalLongitude'], errors='coerce')

# Drop rows where conversion failed
occ = occ.dropna(subset=['decimalLatitude', 'decimalLongitude'])

# Create a GeoDataFrame from the DataFrame
geometry = [Point(xy) for xy in zip(occ['decimalLongitude'], occ['decimalLatitude'])]
gdf = gpd.GeoDataFrame(occ, geometry=geometry)

# Keep unique geometries
gdf = gdf.drop_duplicates(subset=['geometry'])
```

```python
gdf.to_file("/data/cher/Sat2Habitat/data/occurrence.shp")
```
